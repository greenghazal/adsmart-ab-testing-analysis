<font face="vazir" size=3>
نمودار میله ای (Bar Plot) زیر تعداد کاربران قرار گرفته در هر گروه ad یا psa را به نمایش می گذارد که می توانید فراوانی این دو گروه را با یکدیگر مقایسه کنید.

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<span style="color:green"><b>نکته:</b></span>
برخی از نمودارهای این تمرین به کمک کتابخانه <code>Plotly</code> رسم شده است. plotly به شما کمک میکند تا چارت های تعاملی رسم کنید و در ارائه گزارش ها و داشبوردهای تحلیل داده ای یاور شما خواهد بود. برای مثال اگر روی نمودارهای زیر نشانگر موس خود را قرار دهید فراوانی هر دسته را خواهید دید.
</font>
</p>


```python
group_counts = df['test group'].value_counts().reset_index()
group_counts.columns = ['Group', 'Number of Users']


fig = px.bar(
    group_counts,
    x='Group',
    y='Number of Users',
    title='<b>Distribution of Users in Test Groups</b>', # عنوان با تگ html برای bold شدن
    labels={'Group': 'Test Group', 'Number of Users': 'Count of Users'},
    text_auto=True, # نمایش خودکار مقادیر روی میله‌ها
    color='Group', # رنگ‌آمیزی بر اساس گروه
    color_discrete_map={ # تعیین رنگ دلخواه برای هر گروه
        'ad': '#4C72B0',
        'psa': '#55A868'
    }
)


fig.update_layout(
    xaxis_title='Group',
    yaxis_title='Number of Users',
    title_x=0.5, # وسط‌چین کردن عنوان
    font=dict(
        family="Arial, sans-serif",
        size=14,
        color="RebeccaPurple"
    )
)

fig.show()
```

```python

group_counts = df['test group'].value_counts().reset_index()
group_counts.columns = ['Group', 'Count']
# رسم نمودار دایره‌ای تعاملی
fig = px.pie(
    group_counts,
    names='Group',
    values='Count',
    title='<b>User Distribution by Test Group</b>',
    color='Group', # رنگ‌بندی بر اساس نام گروه
    color_discrete_map={ # تعیین رنگ دلخواه برای هر گروه
        'ad': '#4C72B0',
        'psa': '#55A868'
    },
    hole=0.3 # ایجاد یک نمودار دونات برای زیبایی بیشتر (اختیاری)
)

# سفارشی‌سازی جزئیات
fig.update_traces(
    textposition='inside',
    textinfo='percent+label', # نمایش درصد و نام گروه
    pull=[0, 0.1] # جدا کردن قطعه دوم برای تاکید
)
fig.update_layout(
    title_x=0.5, # وسط‌چین کردن عنوان
    legend_title_text='Groups',
    font=dict(family="Arial, sans-serif", size=14)
)

fig.show()
```

<section dir="rtl" style="direction: rtl; text-align: justify; line-height: 200%; font-family: Vazir; font-size: medium;">

<p>
شاید اینکه نسبت دو گروه <strong>۵۰/۵۰</strong> نیست، یا حتی نزدیک به این اعداد هم نیست، عجیب و غیرمنطقی به نظر بیاید و از خود بپرسید:
<span style="color: #006699;"><strong>«چرا برای تست یک فرضیه و نتیجه‌گیری، این‌قدر اختلاف بین فراوانی افراد دو گروه وجود دارد؟»</strong></span>
اما در دنیای کسب‌وکار، این مقیاس و این نحوه‌ی عملکرد، کاملاً <strong>منطقی</strong> است.
</p>

<p>
علت این انتخاب، یک تصمیم <strong>کاملاً آگاهانه و تجاری</strong> است، نه یک اشتباه.
اکثریت مردم در معرض <strong>تبلیغات</strong> قرار خواهند گرفت (گروه <strong>آزمایش</strong>)، و بخش کوچکی از مردم (گروه <strong>کنترل</strong>) به جای آن، یک <strong>اعلان خدمات عمومی (PSA)</strong> خواهند دید. چرا؟
</p>

<p>
گروه PSA (اعلان خدمات عمومی) ذاتاً یک گروه <span style="color: #cc0000;"><strong>بدون درآمد</strong></span> است. این گروه فقط برای ایجاد یک <strong>خط پایه‌ی علمی</strong> جهت مقایسه وجود دارد. هیچ شرکتی حاضر نیست <strong>۵۰٪</strong> از کاربران خود را در گروهی قرار دهد که از پیش می‌داند <strong>هیچ سودی برایش نخواهد داشت</strong>.

این گروه فقط یک گروه کنترل است تا مشخص شود:
<span style="color: #990099;">«اگر تبلیغی وجود نداشت، رفتار کاربران چگونه بود؟»</span>  
برای به‌دست آوردن این معیار، نیازی به اختصاص نیمی از کاربران نیست. یک <strong>نمونه‌ی کوچک اما آماری معتبر</strong> (که در این دیتاست با هزاران کاربر، <strong>۴٪</strong> کاملاً کافی است)، می‌تواند این خط پایه را با دقت بالا مشخص کند.
</p>

<p>
بنابراین، منطق شرکت این است:
</p>

<ul>
  <li><strong>ما می‌خواهیم با تبلیغ جدید (ad) پول دربیاوریم.</strong></li>
  <li><strong>برای اینکه بفهمیم چقدر پول درآوردیم، نیاز به یک گروه مقایسه (psa) داریم که پولی درنمی‌آورد.</strong></li>
  <li><strong>پس برای به حداکثر رساندن سود بالقوه در حین آزمایش، گروه بدون سود (psa) را تا حد ممکن کوچک نگه می‌داریم و گروهی که شانس سودآوری دارد (ad) را تا حد ممکن بزرگ می‌کنیم.</strong></li>
</ul>

<p>
پس این نسبت نامتقارن، هسته‌ی اصلی یک استراتژی هوشمندانه A/B تست در دنیای واقعی است که هم اعتبار آماری را حفظ می‌کند و هم اهداف تجاری را در اولویت قرار می‌دهد.
</p>
</section>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
نرخ تبدیل (Conversion Rate) درصدی از کاربران است که یک اقدام یا هدف مشخص مورد نظر شما را تکمیل می‌کنند؛ مانند خرید محصول، ثبت‌نام در سایت یا کلیک روی یک لینک. این شاخص کلیدی، میزان موفقیت یک کمپین تبلیغاتی یا صفحه وب را در رسیدن به هدفش اندازه‌گیری می‌کند.
</font>
</p>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در سلول زیر نرخ تبدیل را برای کل کاربران محاسبه کنید. یعنی بطور کلی چند درصد کاربران محصول را خریده اند؟
</font>
</p>


```python
num_converted=df['converted'].sum()
total_users=len(df)
conversion_rate=(num_converted/total_users)*100
print(conversion_rate)
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
یک نمودار میله‌ای برای مقایسه بصری نرخ تبدیل (Conversion Rate) در دو گروه ad و psa ترسیم کنید.
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<span style="color:green"><b>نکته:</b></span>
برای رسم نمودارهای این تمرین، اینکه از کدام کتابخانه استفاده شده یا جزئیات نمودارها مثل title و... خیلی اهمیتی ندارد مهم این است خواسته مسئله بدرستی اجابت شود و مقادیر نهایی رسم شده درست باشند. اما سعی کنید تا حد امکان نمودارهای جذاب و خوانا ترسیم کنید.
</font>
</p>



```python
conversion_df=pd.DataFrame({'group':['ad','psa'],'conversion_rate':[df[df['test group']=='ad']['converted'].mean()*100,df[df['test group']=='psa']['converted'].mean()*100]})
plt.figure(figsize=(10,6))
plt.bar(conversion_df['group'],conversion_df['conversion_rate'])
plt.title('conversion rate by gtoup')
plt.ylabel('%')
plt.show()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
توزیع ستون total_ads (تعداد کل تبلیغات مشاهده شده) را با استفاده از یک نمودار هیستوگرام (Histogram) مشاهده میکنید.
</font>

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# --- رسم هیستوگرام ---
ax = sns.histplot(
    data=df,
    x='total ads',
    bins=50,
    kde=True, # افزودن منحنی تخمین چگالی
    color='#55A868'
)

# --- افزودن عنوان و لیبل‌ها ---
ax.set_title('Distribution of Total Ads Seen (Seaborn)', fontsize=16, weight='bold')
ax.set_xlabel('Number of Ads Seen', fontsize=12)
ax.set_ylabel('Frequency (Number of Users)', fontsize=12)

plt.show()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
توزیع ستون total_ads (تعداد کل تبلیغات مشاهده شده) را با استفاده از یک نمودار جعبه ای (Box Plot) نمایش دهید. خروجی ها را با یکدیگر مقایسه کرده همچنین رسم نمودار جعبه ای به شناسایی داده های پرت کمک شایان تری میکند.

<font face="vazir" size=3>
با استفاده از یک نمودار مناسب (مانند نمودار میله‌ای گروه‌بندی شده)، بررسی کنید که آیا روز هفته‌ای که بیشترین تبلیغ نمایش داده شده (most_ads_day)، بر روی نرخ تبدیل تأثیرگذار بوده است یا خیر.

برای تحلیل دقیق‌تر تأثیر تعداد تبلیغات، کاربران را بر اساس ستون total_ads به چند گروه (bin) تقسیم کنید (مثلاً گروه ۱-۵ تبلیغ، ۶-۱۰ تبلیغ، و...). سپس نرخ تبدیل را برای هر گروه محاسبه و با یک نمودار میله‌ای (Bar Chart) نمایش دهید. این کار نشان می‌دهد که آیا با افزایش تعداد تبلیغات، نرخ تبدیل نیز افزایش می‌یابد یا خیر.

<h2 align="right" dir="rtl" style="line-height: 200%; font-family: vazir; color: #0099cc; font-size: x-large;">
  آزمون فرضیه آماری (Hypothesis Testing)
</h2>

<p dir="rtl" style="direction: rtl; text-align: justify; line-height: 200%; font-family: vazir; font-size: medium;">
  فرضیه‌های آماری خود را تعریف کنید:
</p>

<ul dir="rtl" style="font-family: vazir; font-size: medium; line-height: 200%; text-align: justify;">
  <li>
    <strong>فرضیه صفر (H<sub>0</sub>)</strong>: نرخ تبدیل گروه (<code>ad</code>) و گروه (<code>psa</code>) برابر است. این فرضیه بیانگر عدم تأثیرگذاری تبلیغ جدید است.
  </li>
  <li>
    <strong>فرضیه جایگزین (H<sub>1</sub>)</strong>:نرخ تبدیل گروه (<code>ad</code>) بیشتر از گروه (<code>psa</code>) است. این فرضیه بیانگر موفقیت‌آمیز بودن کمپین تبلیغاتی است.
</ul>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای انجام آزمون آماری، یک جدول توافقی (Contingency Table) با استفاده از تابع pd.crosstab بسازید. جدولی بسازید که تعداد کاربران تبدیل شده (True) و تبدیل نشده (False) را به تفکیک برای هر گروه (ad و psa) نمایش دهد.
</font>
</p>


```python
continengency=pd.crosstab(df['test group'],df['converted'])
print(continengency)
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
با استفاده از جدول توافقی ساخته شده و <a href="https://www.statsmodels.org/dev/generated/statsmodels.stats.proportion.proportions_ztest.html">آزمون برابری دو نسبت (Test for Equality of Two Proportions)</a> استفاده کنید و ببینید آیا تفاوت نرخ تبدیل بین گروه ad و psa به اندازه‌ای بزرگ است که بتوانیم آن را معنادار بدانیم. خروجی‌های اصلی آن، یعنی آماره Z و مقدار p-value را چاپ کنید.
</font>
</p>


```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
truess=[14423,420]
totals=[550154+14423,23104+420]
z_stat,p_value=proportions_ztest(truess,totals,alternative='larger')
print(f'z:{z_stat:.2f}',f'p_value: {p_value:.2f}')
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
نتیجه گیری خود را از انجام این آزمون در یک یا دو خط بنویسید.
</font>
</p>


<p dir=rtl style="direction: rtl;text-align: center;line-height:200%;font-family:vazir;font-size:medium;color:#0099cc"><font face="vazir" size=3><i>
نتیجه آزمون نشون داد که مقدار p-value  صفره و این یعنی اختلاف بین دو گروه شانسی نیست  
بنابراین فرض صفر که میگفت نرخ تبدیل دو گروه برابره رو نمی‌تونیم قبول کنیم و رد میشه  
در نتیجه فرض مقابل تأیید میشه و میشه گفت تبلیغ جدید باعث افزایش نرخ تبدیل شده
</i></font></p>


<p dir="rtl" style="direction: rtl; text-align: justify; line-height: 200%; font-family: vazir; font-size: medium;">
  این سؤال به بررسی ارتباط بین دو متغیر کیفی <code>most_ads_day</code> (روزی از هفته که کاربر بیشترین تبلیغ را دیده) و <code>converted</code> (تبدیل‌شدن یا نشدن کاربر) می‌پردازد.
</p>

<h3 align="right" dir="rtl" style="line-height: 200%; font-family: vazir; color: #006699; font-size: medium;">
  مراحل تحلیل:
</h3>

<ol dir="rtl" style="font-family: vazir; font-size: medium; line-height: 200%; text-align: justify;">
  <li><strong>ساخت جدول توافقی (Contingency Table):</strong> شمارش فراوانی هر ترکیب از دو متغیر</li>
  <li><strong>اجرای آزمون کای-دو (Chi<sup>2</sup>):</strong> بررسی وابستگی یا استقلال این دو متغیر</li>
</ol>

<h3 align="right" dir="rtl" style="line-height: 200%; font-family: vazir; color: #006699; font-size: medium;">
  فرضیه‌ها:
</h3>

<ul dir="rtl" style="font-family: vazir; font-size: medium; line-height: 200%; text-align: justify;">
  <li><strong>فرضیه صفر (H<sub>0</sub>):</strong> بین روز هفته و تبدیل شدن کاربران ارتباطی وجود ندارد (دو متغیر مستقل‌اند).</li>
  <li><strong>فرضیه جایگزین (H<sub>1</sub>):</strong> بین این دو متغیر ارتباط وجود دارد.</li>
</ul>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
نتیجه آزمون را تفسیر کنید. با در نظر گرفتن سطح معناداری 
alpha=0.05
، آیا فرضیه صفر رد می‌شود؟ نتیجه‌گیری نهایی خود را در مورد اینکه آیا کمپین تبلیغاتی جدید به طور معناداری مؤثر بوده است یا خیر، در چند جمله بنویسید.
</font>
</p>


```python
continengency=pd.crosstab(df['most ads day'],df['converted'])
print(continengency)
```

```python
from scipy.stats import chi2_contingency
```

```python
stat,p,dof,expected=chi2_contingency(continengency)
print(f'chi_2 stat:{stat:.2f}',f'p value :{p:.2f}')
alpha=0.05
if p <=alpha:
    print('reject h0')
else:
    print('h0 accepted')
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
اگر فرضیه صفر را رد کردید؟ میتوانید بگویید بهترین روز برای تبلیغ چه روزی است؟ (پاسخ در markdown نوشته شود.)
</font>
</p>


<p dir=rtl style="direction: rtl;text-align: center;line-height:200%;font-family:vazir;font-size:medium;color:#0099cc"><font face="vazir" size=3><i>
نتیجه آزمون کای دو نشون داد مقدار آماره 410.05 شده و p-value تقریباً صفره که خیلی کوچیک‌تر از سطح معناداری 0.05 حساب میشه.  
به خاطر همین فرضیه صفر رد میشه و می‌تونیم بگیم بین روز هفته و تبدیل شدن کاربرا یه ارتباط واقعی وجود داره.  
یعنی اینکه روزی که تبلیغ نمایش داده میشه روی احتمال خرید کاربر تاثیر داره و همه روزها نتیجه یکسانی ندارن.
اگه به جدول نگاه کنیم می‌بینیم روز Monday بیشترین تعداد تبدیل رو داشته.  
پس دوشنبه بهترین روز برای نمایش تبلیغ بوده </i></font></p>

<h2 align="right" dir="rtl" style="line-height: 200%; font-family: vazir; color: #0099cc; font-size: x-large;">
  تحلیل تأثیر تعداد تبلیغات بر فروش
</h2>

<p dir="rtl" style="direction: rtl; text-align: justify; line-height: 200%; font-family: vazir; font-size: medium;">
  یک سوال تجاری بسیار مهم این است که آیا بین تعداد تبلیغاتی که یک کاربر می‌بیند و اینکه آیا نهایتاً خرید می‌کند، ارتباطی وجود دارد؟ به عبارت دیگر، آیا کاربرانی که خرید کرده‌اند، به طور میانگین تعداد تبلیغات متفاوتی دیده‌اند؟
</p>

<h3 align="right" dir="rtl" style="line-height: 200%; font-family: vazir; color: #006699; font-size: medium;">
  فرضیه‌ها:
</h3>

<ul dir="rtl" style="font-family: vazir; font-size: medium; line-height: 200%; text-align: justify;">
  <li><strong>فرضیه صفر (H<sub>0</sub>):</strong> میانگین تعداد تبلیغات مشاهده شده توسط کاربران تبدیل شده و نشده، یکسان است.</li>
  <li><strong>فرضیه جایگزین (H<sub>1</sub>):</strong> میانگین تعداد تبلیغات مشاهده شده توسط این دو گروه متفاوت است.</li>
</ul>

<h3 align="right" dir="rtl" style="line-height: 200%; font-family: vazir; color: #006699; font-size: medium;">
  خواسته:
</h3>

<p dir="rtl" style="direction: rtl; text-align: justify; line-height: 200%; font-family: vazir; font-size: medium;">
برای سنجش این فرضیه، <strong>میانگین</strong> ستون <code>total_ads</code> را برای هر دو گروه (کاربران تبدیل شده و نشده) مقایسه کنید. با اجرای آزمون آماری مناسب، بررسی کنید که آیا تفاوت مشاهده شده بین این دو میانگین، از نظر آماری معنادار است یا خیر و p-value حاصل را گزارش دهید.
</p>

```python
from scipy.stats import ttest_ind
```

```python
ads_converted=df[df['converted']==True]['total ads']
ads_not_converted=df[df['converted']==False]['total ads']
t,p=ttest_ind(ads_converted,ads_not_converted,equal_var=False)
print(f't:{t:.2f},p:{p:.2f}')
```

<p dir=rtl style="direction: rtl;text-align: center;line-height:200%;font-family:vazir;font-size:medium;color:#0099cc"><font face="vazir" size=3><i>
t شد 84.18 و p تقریبا صفر یعنی شواهد کافی داریم که فرض صفر رو رد کنیم، یعنی میانگین تعداد تبلیغ دیده شده برای خریدارها و غیرخریدارها یکسان نیست
 کسایی که خرید کردن تبلیغات بیشتری دیدن و واضحا تعداد تبلیغ روی خرید تاثیر داره
</i></font></p>
